```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('drawndata1.csv')
```

```python
df.head()
```

```python
X = df[['x','y']].values
y = df['z'] == 'a'
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
from sklearn.preprocessing import StandardScaler 
X_new = StandardScaler().fit_transform(X)
plt.scatter(X_new[:,0],X_new[:,1], c=y)
```

![](diagram2.png)

changing standardscaler to quantiles for scaling

```python
from sklearn.preprocessing import QuantileTransformer
X_new = QuantileTransformer(n_quantiles=100).fit_transform(X) # n_quantiles is set to 100 for specific dataset, normally 1000
plt.scatter(X_new[:,0],X_new[:,1], c=y)
```
